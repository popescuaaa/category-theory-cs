### Gas prices correlation problem ~ lab 2

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from statistics import correlation, pvariance
```

```python
# Load data
df = pd.read_csv('./pretcarburant.csv')
```

```python
df.info()
```

```python
# Find basic infor regarding the data
disel = df['Motorina'].values
normal = df['Benzina'].values

print("Min / Max / Mean disel: {} / {} / {}".format(np.min(disel), np.max(disel), np.mean(disel)))
print("Min / Max / Mean normal: {} / {} / {}".format(np.min(normal), np.max(normal), np.mean(normal)))
```

```python
# Create a small linear plot for both prices to see time evolution
fig, ax = plt.subplots(1, 1)
ax.plot(disel, label="Disel")
ax.plot(normal, label="Normal")
ax.legend(loc='right')
```

```python
print("Dispersion disel: {}".format(pvariance(disel)))
print("Dispersion normal: {}".format(pvariance(normal)))
print("Correlation {}".format(correlation(disel, normal)))
```

#### Very correlated as shown above and with higher precision in the data aquisition process

```python
lr = LinearRegression(fit_intercept=True, n_jobs=10)
lr.fit(np.array([i for i in range(len(disel))]).reshape(-1, 1), disel)
```

```python
predictions = lr.predict(disel.reshape(-1, 1))
```

```python
# Compare
fig, ax = plt.subplots(1, 1)
ax.plot(disel, label="Disel - real")
ax.plot(predictions, label="Disel - predicted")
ax.legend(loc='right')
```
